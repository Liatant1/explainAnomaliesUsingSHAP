```python
import numpy as np
import pandas as pd
```

```python
n_records = 1000000
n_features = 4

features_names = []
for feature in range(1, n_features+1):
    features_names.append('feature_' + str(feature))

data = np.random.rand(n_records, n_features)
```

```python
data_df = pd.DataFrame(data=data, columns=features_names)
```

```python
data_df.head()
```

```python
data_df['feature_1_2'] = data_df['feature_1'] + data_df['feature_2']
data_df['feature_3_4'] = data_df['feature_3'] + data_df['feature_4']

data_df['class'] = np.zeros(n_records)
```

```python
data_df.head()
```

```python
n_to_change = 15000

idx = data_df.sample(n=n_to_change).index

for i in idx:
#   rand_vals = np.random.rand(2)
#   data_df.loc[i, 'feature_1_2'] = rand_vals[0]
#   data_df.loc[i, 'feature_3_4'] = rand_vals[1]
#   data_df.loc[i, 'class'] = '1'
    rand_num = np.random.rand(1)[0]
    if rand_num < 0.5:
        data_df.loc[i, 'feature_1_2'] = np.random.rand(1)[0]
        data_df.loc[i, 'class'] = '1_2'
    else:
        data_df.loc[i, 'feature_3_4'] = np.random.rand(1)[0]
        data_df.loc[i, 'class'] = '3_4'
    
```

```python
data_df.head(10)
```

```python
x_max_val = data_df.iloc[:,:-1].max().max()
print(x_max_val)
data_df.iloc[:,:-1] = data_df.iloc[:,:-1] / x_max_val
```

```python
data_df.to_csv('./random_data_' + str(n_features+2) + '_features.csv', index=False)
```
